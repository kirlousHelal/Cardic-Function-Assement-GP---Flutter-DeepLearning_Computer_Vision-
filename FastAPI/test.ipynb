```python
import os
import base64

def get_image_encoded(path):
    with open(path, 'rb') as f:
        image_data = f.read()
        print(type(image_data))
        print(image_data)
    encoded_image = base64.b64encode(image_data).decode('utf-8')
    return encoded_image

img = get_image_encoded('images/ES_frame.jpg')
print(type(img))
print(img)
```
